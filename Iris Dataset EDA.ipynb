```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris=load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target
```

```python
print(data.head())
print(data.describe())
```

```python
import matplotlib.pyplot as plt
data.hist(figsize=(8, 8))
plt.suptitle("Feature Distributions")
plt.show()
```

```python
import seaborn as sns
#Plot the box plots for each feature grouped by species
sns.boxplot(x='species', y='sepal length (cm)', data=data)
plt.title('Sepal Length Distribution by Species')
plt.show()

sns.boxplot(x='species', y='sepal width (cm)', data=data)
plt.title('Sepal Width Distribution by Species')
plt.show()

sns.boxplot(x='species', y='petal width (cm)', data=data)
plt.title('Petal Width Distribution by Species')
plt.show()

sns.boxplot(x='species', y='petal width (cm)', data=data)
plt.title('Petal Width Distribution by Species')
plt.show()
```

```python
sns.pairplot(data, hue='species', palette='Set1')
plt.show()
```

```python
#Correlation Matrix
corr_matrix = data.iloc[:, :-1].corr()
print('Correlation Matrix:')
print(corr_matrix)
```

```python
#Correlation Matrix Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Iris Features')
plt.show()
```
